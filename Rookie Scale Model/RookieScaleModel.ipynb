```python
import pandas as pd
import statsmodels.api as sm
from google.colab import files

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

selected_features = ['orb', 'drb', 'trb', 'stl', 'blk', 'usg', 'vorp', 'ppg']
X_train = df[selected_features]


X_train = X_train.apply(pd.to_numeric, errors='coerce')

X_train = sm.add_constant(X_train)

y_train = df['cap_percentage']

y_train = pd.to_numeric(y_train, errors='coerce')

rookiescalermodel = sm.OLS(y_train, X_train).fit()

print(rookiescalermodel.summary())
```

```python
import pandas as pd
import locale

locale.setlocale(locale.LC_ALL, '')

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
test_data = pd.read_csv(file_name)

X_test = test_data[['orb', 'drb', 'trb', 'stl', 'blk', 'usg', 'vorp', 'ppg']]

X_test = X_test.fillna(X_test.mean())

X_test = sm.add_constant(X_test)  # Add constant if your model was trained with a constant
predictions = rookiescalermodel.predict(X_test)

test_data['percentage of cap'] = predictions
test_data['salary'] = predictions * 140588000  # Multiply predictions as needed
test_data['salary'] = test_data['salary'].apply(lambda x: locale.format_string("%.2f", x, grouping=True))

test_data.head(25)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

actual_values = test_data['cap_percentage']
predictions = test_data['percentage of cap']

mae = mean_absolute_error(actual_values, predictions)
mse = mean_squared_error(actual_values, predictions)
rmse = mean_squared_error(actual_values, predictions, squared=False)  # Calculate RMSE
r2 = r2_score(actual_values, predictions)



print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2 score): {r2}")
```
